# Regresion Lineal con Jupyter Notebook - DATA-DOGS

Notebook alineado con la guia del curso para construir:
- EDA completo
- Regresion lineal simple
- Regresion lineal multiple
- Diagnosticos estadisticos (normalidad, homocedasticidad, multicolinealidad)

Variables del modelo:
- Variable dependiente / objetivo (Y): `vida_promedio`
- Variables independientes (X) - modelo simple: `peso_promedio`
- Variables independientes (X) - modelo multiple: `peso_promedio`, `diferencia_sexual_peso`, `hipo_num`
- Variables objetivo transformadas para diagnostico: `log(vida_promedio)`, `sqrt(vida_promedio)`


```python
# CELDA 1 - Importaciones y configuracion
# 🎯 Objetivo: importar librerias y configurar estilo para todo el notebook
import os
import sys

sys.path.insert(0, '..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.outliers_influence import variance_inflation_factor

from scripts.database import engine

os.makedirs('../data/graficas', exist_ok=True)

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 120, 'font.size': 11})

print('OK - Librerias importadas')
```

```python
# CELDA 2 - Carga de datos desde PostgreSQL
# 🎯 Objetivo: cargar datos desde PostgreSQL en un DataFrame para el analisis
QUERY = """
SELECT
    raza,
    vida_min,
    vida_max,
    vida_promedio,
    peso_macho_max,
    peso_hembra_max,
    peso_promedio,
    categoria_peso,
    diferencia_sexual_peso,
    hipoalergenico,
    fecha_extraccion
FROM razas_perros
ORDER BY fecha_extraccion
"""

df = pd.read_sql(QUERY, engine)

print('OK - Datos cargados')
print(f'Filas    : {df.shape[0]:,}')
print(f'Columnas : {df.shape[1]}')
df.head()
```

```python
# CELDA 3 - Inspeccion inicial
# 🎯 Objetivo: inspeccionar estructura, nulos y categorias del dataset
print('=' * 60)
print('INFORMACION GENERAL DEL DATASET')
print('=' * 60)
print(df.info())

print()
print('Nulos por columna:')
print(df.isnull().sum())

print()
print('Categorias de peso:')
print(df['categoria_peso'].value_counts(dropna=False))
```

```python
# CELDA 4 - Estadisticas descriptivas
# 🎯 Objetivo: resumir estadisticas descriptivas de variables numericas
vars_num = [
    'vida_promedio',
    'peso_promedio',
    'peso_macho_max',
    'peso_hembra_max',
    'diferencia_sexual_peso',
]

print('ESTADISTICAS DESCRIPTIVAS')
print(df[vars_num].describe().round(3))
```

```python
# CELDA 5 - Histogramas de distribucion
# 🎯 Objetivo: visualizar distribuciones para detectar sesgos y dispersion
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribucion de Variables - DATA-DOGS', fontsize=14, fontweight='bold', y=1.02)

datos = {
    'vida_promedio': ('Vida promedio (anos)', '#4ecdc4'),
    'peso_promedio': ('Peso promedio (kg)', '#ff6b6b'),
    'peso_macho_max': ('Peso macho max (kg)', '#ffa07a'),
    'diferencia_sexual_peso': ('Diferencia sexual peso (kg)', '#95e1d3'),
}

for ax, (col, (label, color)) in zip(axes.flat, datos.items()):
    ax.hist(df[col].dropna(), bins=35, color=color, edgecolor='white', alpha=0.85)
    media = df[col].mean()
    ax.axvline(media, color='#2a5298', linestyle='--', linewidth=1.8, label=f'Media: {media:.2f}')
    ax.set_title(label, fontweight='bold')
    ax.set_xlabel(label)
    ax.set_ylabel('Frecuencia')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../data/graficas/eda_distribuciones_dogs.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CELDA 6 - Boxplots por categoria de peso
# 🎯 Objetivo: comparar distribuciones por categoria de peso con boxplots
orden = ['Muy pequeno', 'Pequeno', 'Mediano', 'Grande', 'Gigante']
orden_disponible = [c for c in orden if c in df['categoria_peso'].dropna().unique()]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(data=df, x='categoria_peso', y='peso_promedio', order=orden_disponible, palette='Set2', ax=axes[0])
axes[0].set_title('Peso promedio por categoria', fontweight='bold')
axes[0].set_xlabel('')
axes[0].tick_params(axis='x', rotation=20)

sns.boxplot(data=df, x='categoria_peso', y='vida_promedio', order=orden_disponible, palette='Set3', ax=axes[1])
axes[1].set_title('Vida promedio por categoria', fontweight='bold')
axes[1].set_xlabel('')
axes[1].tick_params(axis='x', rotation=20)

plt.tight_layout()
plt.savefig('../data/graficas/eda_boxplots_categoria_dogs.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CELDA 7 - Matriz de correlacion
# 🎯 Objetivo: medir correlaciones entre variables para guiar el modelado
corr = df[vars_num].corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,
    fmt='.3f',
    cmap='coolwarm',
    center=0,
    linewidths=0.6,
    square=True,
    cbar_kws={'shrink': 0.8}
)
plt.title('Matriz de Correlacion - DATA-DOGS', fontsize=13, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('../data/graficas/eda_correlacion_dogs.png', dpi=150, bbox_inches='tight')
plt.show()

print()
print('Correlacion con vida_promedio:')
print(corr['vida_promedio'].drop('vida_promedio').sort_values(ascending=False).round(4))
```

```python
# CELDA 8 - Scatter plots bivariados
# 🎯 Objetivo: evaluar relaciones lineales con scatter y tendencia
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Relacion de Features con Vida Promedio', fontsize=13, fontweight='bold')

pares = [
    ('peso_promedio', 'Peso promedio (kg)', '#ff6b6b'),
    ('peso_macho_max', 'Peso macho max (kg)', '#ffa07a'),
    ('diferencia_sexual_peso', 'Diferencia sexual peso (kg)', '#95e1d3'),
]

for ax, (feature, xlabel, color) in zip(axes, pares):
    sub = df[[feature, 'vida_promedio']].dropna()
    ax.scatter(sub[feature], sub['vida_promedio'], alpha=0.25, s=15, color=color)
    z = np.polyfit(sub[feature], sub['vida_promedio'], 1)
    p = np.poly1d(z)
    xp = np.linspace(sub[feature].min(), sub[feature].max(), 200)
    ax.plot(xp, p(xp), color='#2a5298', linewidth=2)
    r = sub.corr().iloc[0, 1]
    ax.set_title(f'r = {r:.3f}', fontweight='bold')
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Vida promedio (anos)')
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../data/graficas/eda_scatter_dogs.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CELDA 9 - Train/Test para modelo simple
# 🎯 Objetivo: separar datos en entrenamiento y prueba
X_simple = df[['peso_promedio']].values
y = df['vida_promedio'].values

X_train, X_test, y_train, y_test = train_test_split(
    X_simple, y, test_size=0.20, random_state=42
)

print(f'Train: {X_train.shape[0]:,} muestras ({X_train.shape[0]/len(y)*100:.0f}%)')
print(f'Test : {X_test.shape[0]:,} muestras ({X_test.shape[0]/len(y)*100:.0f}%)')
```

```python
# CELDA 10 - Entrenamiento regresion simple
# 🎯 Objetivo: entrenar el modelo de regresion lineal simple
modelo_simple = LinearRegression()
modelo_simple.fit(X_train, y_train)

y_pred_simple = modelo_simple.predict(X_test)

print(f'Intercepto  b0 : {modelo_simple.intercept_:.4f}')
print(f'Coeficiente b1 : {modelo_simple.coef_[0]:.4f}')
print(f'Ecuacion: vida_promedio = {modelo_simple.intercept_:.3f} + ({modelo_simple.coef_[0]:.3f}) * peso_promedio')
```

```python
# CELDA 11 - Visualizacion de recta de regresion simple
# 🎯 Objetivo: mostrar la recta ajustada frente a datos reales
fig, ax = plt.subplots(figsize=(10, 6))

ax.scatter(X_test, y_test, alpha=0.35, s=18, color='#4ecdc4', label='Datos observados (test)')
x_line = np.linspace(X_simple.min(), X_simple.max(), 300).reshape(-1, 1)
ax.plot(x_line, modelo_simple.predict(x_line), color='#ff6b6b', linewidth=2.5, label='Recta de regresion')

ax.set_xlabel('Peso promedio (kg)', fontsize=12)
ax.set_ylabel('Vida promedio (anos)', fontsize=12)
ax.set_title('Regresion Lineal Simple: Peso Promedio -> Vida Promedio', fontsize=13, fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../data/graficas/regresion_simple_dogs.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CELDA 12 - Metricas modelo simple
# 🎯 Objetivo: calcular metricas base del modelo simple
r2 = r2_score(y_test, y_pred_simple)
mse = mean_squared_error(y_test, y_pred_simple)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_simple)

print('=' * 52)
print('📈 METRICAS - REGRESION LINEAL SIMPLE')
print('=' * 52)
print(f'🎯 R2   : {r2:.4f} ({r2*100:.2f}%)')
print(f'📉 MSE  : {mse:.4f}')
print(f'📏 RMSE : {rmse:.4f}')
print(f'🧮 MAE  : {mae:.4f}')
print('=' * 52)
```

```python
# CELDA 13 - Resumen OLS modelo simple
# 🎯 Objetivo: obtener inferencia estadistica OLS del modelo simple
X_ols = sm.add_constant(df['peso_promedio'])
modelo_ols_simple = sm.OLS(df['vida_promedio'], X_ols).fit()
print(modelo_ols_simple.summary())
```

```python
# CELDA 14 - Normalidad de residuos (simple)
# 🎯 Objetivo: verificar normalidad de residuos del modelo simple
y_pred_all = modelo_simple.predict(df[['peso_promedio']].values)
residuos = df['vida_promedio'].values - y_pred_all

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Supuesto de Normalidad - Modelo Simple', fontsize=13, fontweight='bold')

axes[0].hist(residuos, bins=40, color='#4ecdc4', edgecolor='white', density=True, alpha=0.85)
xr = np.linspace(residuos.min(), residuos.max(), 200)
axes[0].plot(xr, stats.norm.pdf(xr, residuos.mean(), residuos.std()), color='#ff6b6b', linewidth=2.5)
axes[0].axvline(0, color='#2a5298', linestyle='--', linewidth=1.5)
axes[0].set_title('Distribucion de residuos', fontweight='bold')
axes[0].set_xlabel('Residuos (anos)')
axes[0].set_ylabel('Densidad')
axes[0].grid(axis='y', alpha=0.3)

(osm, osr), (slope, intercept, _) = stats.probplot(residuos, dist='norm')
axes[1].scatter(osm, osr, alpha=0.4, s=12, color='#4ecdc4')
axes[1].plot(osm, slope * np.array(osm) + intercept, color='#ff6b6b', linewidth=2.5)
axes[1].set_title('Q-Q plot de residuos', fontweight='bold')
axes[1].set_xlabel('Cuantiles teoricos')
axes[1].set_ylabel('Cuantiles observados')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../data/graficas/normalidad_simple_dogs.png', dpi=150, bbox_inches='tight')
plt.show()

muestra_sw = residuos if len(residuos) <= 5000 else np.random.default_rng(42).choice(residuos, 5000, replace=False)
stat_sw, p_sw = stats.shapiro(muestra_sw)

print()
print('Test de Shapiro-Wilk:')
print(f'W       : {stat_sw:.4f}')
print(f'p-value : {p_sw:.6f}')
print('Conclusion:', 'No se rechaza normalidad' if p_sw > 0.05 else 'Se rechaza normalidad')
```

```python
# CELDA 15 - Homocedasticidad (simple)
# 🎯 Objetivo: evaluar homocedasticidad con residuos y Breusch-Pagan
residuos_ols = modelo_ols_simple.resid
ajustados_ols = modelo_ols_simple.fittedvalues

plt.figure(figsize=(10, 5))
plt.scatter(ajustados_ols, residuos_ols, alpha=0.30, s=15, color='#4ecdc4')
plt.axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess = sm.nonparametric.lowess(residuos_ols, ajustados_ols, frac=0.3)
plt.plot(lowess[:, 0], lowess[:, 1], color='#ffa07a', linewidth=2.5, label='LOWESS')
plt.xlabel('Valores ajustados (y-hat)')
plt.ylabel('Residuos')
plt.title('Homocedasticidad: residuos vs ajustados', fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('../data/graficas/homoc_simple_dogs.png', dpi=150, bbox_inches='tight')
plt.show()

lm, lm_p, fval, fp = het_breuschpagan(residuos_ols, modelo_ols_simple.model.exog)

print()
print('Test de Breusch-Pagan:')
print(f'LM      : {lm:.4f}')
print(f'p-value : {lm_p:.6f}')
print('Conclusion:', 'Homocedasticidad OK' if lm_p > 0.05 else 'Heterocedasticidad detectada')
```

## Ajustes de Y para heterocedasticidad (log y raiz)
A continuacion se replica el flujo completo del modelo simple usando dos transformaciones de la variable objetivo `Y = vida_promedio`.


```python
# CELDA 15A - Transformacion logaritmica de la variable objetivo
# 🎯 Objetivo: crear transformacion logaritmica de la variable objetivo
df['y_log'] = np.log(df['vida_promedio'])

print('OK - Se creo y_log = log(vida_promedio)')
print(df[['vida_promedio', 'y_log']].head())
```

```python
# CELDA 15B - Regresion simple con transformacion logaritmica
# 🎯 Objetivo: entrenar regresion simple usando log(Y)
X_log = df[['peso_promedio']].values
y_log = df['y_log'].values

X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_log, y_log, test_size=0.20, random_state=42
)

modelo_log = LinearRegression()
modelo_log.fit(X_train_log, y_train_log)
y_pred_log = modelo_log.predict(X_test_log)

print(f'Intercepto b0 (log): {modelo_log.intercept_:.4f}')
print(f'Coeficiente b1 (log): {modelo_log.coef_[0]:.4f}')
print(f'Ecuacion: log(vida_promedio) = {modelo_log.intercept_:.3f} + ({modelo_log.coef_[0]:.3f}) * peso_promedio')
```

```python
# CELDA 15C - Metricas del modelo con log
# 🎯 Objetivo: evaluar metricas log en escala transformada y original
# Metricas en escala transformada (log)
r2_log = r2_score(y_test_log, y_pred_log)
mse_log = mean_squared_error(y_test_log, y_pred_log)
rmse_log = np.sqrt(mse_log)
mae_log = mean_absolute_error(y_test_log, y_pred_log)

# Metricas en escala original de vida_promedio (comparables con modelo original)
y_test_log_orig = np.exp(y_test_log)
y_pred_log_orig = np.exp(y_pred_log)
r2_log_orig = r2_score(y_test_log_orig, y_pred_log_orig)
rmse_log_orig = np.sqrt(mean_squared_error(y_test_log_orig, y_pred_log_orig))
mae_log_orig = mean_absolute_error(y_test_log_orig, y_pred_log_orig)

print('=' * 56)
print('METRICAS - REGRESION SIMPLE CON LOG(Y)')
print('=' * 56)
print('Escala log(Y):')
print(f'R2   : {r2_log:.4f} ({r2_log*100:.2f}%)')
print(f'MSE  : {mse_log:.6f}')
print(f'RMSE : {rmse_log:.6f}')
print(f'MAE  : {mae_log:.6f}')
print()
print('Escala original de vida_promedio (comparable):')
print(f'R2   : {r2_log_orig:.4f} ({r2_log_orig*100:.2f}%)')
print(f'RMSE : {rmse_log_orig:.6f}')
print(f'MAE  : {mae_log_orig:.6f}')
print('=' * 56)
```

```python
# CELDA 15D - OLS del modelo con log
# 🎯 Objetivo: obtener resumen OLS del modelo con log(Y)
X_ols_log = sm.add_constant(df['peso_promedio'])
modelo_ols_log = sm.OLS(df['y_log'], X_ols_log).fit()

print(modelo_ols_log.summary())
```

```python
# CELDA 15E - Diagnostico del modelo con log
# 🎯 Objetivo: diagnosticar normalidad y homocedasticidad con log(Y)
y_pred_all_log = modelo_log.predict(df[['peso_promedio']].values)
residuos_log = df['y_log'].values - y_pred_all_log

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Diagnosticos - Modelo con log(Y)', fontsize=13, fontweight='bold')

axes[0].hist(residuos_log, bins=40, color='#4ecdc4', edgecolor='white', density=True, alpha=0.85)
xr = np.linspace(residuos_log.min(), residuos_log.max(), 200)
axes[0].plot(xr, stats.norm.pdf(xr, residuos_log.mean(), residuos_log.std()), color='#ff6b6b', linewidth=2.5)
axes[0].axvline(0, color='#2a5298', linestyle='--')
axes[0].set_title('Distribucion de residuos', fontweight='bold')
axes[0].set_xlabel('Residuos')
axes[0].set_ylabel('Densidad')
axes[0].grid(axis='y', alpha=0.3)

(osm, osr), (s, i, _) = stats.probplot(residuos_log, dist='norm')
axes[1].scatter(osm, osr, alpha=0.35, s=12, color='#4ecdc4')
axes[1].plot(osm, s * np.array(osm) + i, color='#ff6b6b', linewidth=2.5)
axes[1].set_title('Q-Q plot de residuos', fontweight='bold')
axes[1].set_xlabel('Cuantiles teoricos')
axes[1].set_ylabel('Cuantiles observados')
axes[1].grid(alpha=0.3)

axes[2].scatter(modelo_ols_log.fittedvalues, modelo_ols_log.resid, alpha=0.30, s=15, color='#4ecdc4')
axes[2].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_log = sm.nonparametric.lowess(modelo_ols_log.resid, modelo_ols_log.fittedvalues, frac=0.3)
axes[2].plot(lowess_log[:, 0], lowess_log[:, 1], color='#ffa07a', linewidth=2.5, label='LOWESS')
axes[2].set_title('Residuos vs valores ajustados', fontweight='bold')
axes[2].set_xlabel('Valores ajustados')
axes[2].set_ylabel('Residuos')
axes[2].legend()
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../data/graficas/diagnosticos_log_simple_dogs.png', dpi=150, bbox_inches='tight')
plt.show()

muestra_sw_log = residuos_log if len(residuos_log) <= 5000 else np.random.default_rng(42).choice(residuos_log, 5000, replace=False)
stat_sw_log, p_sw_log = stats.shapiro(muestra_sw_log)
lm_log, lm_p_log, _, _ = het_breuschpagan(modelo_ols_log.resid, modelo_ols_log.model.exog)

print('Shapiro-Wilk (log):')
print(f'p-value: {p_sw_log:.6f}')
print('Conclusion:', 'No se rechaza normalidad' if p_sw_log > 0.05 else 'Se rechaza normalidad')
print()
print('Breusch-Pagan (log):')
print(f'p-value: {lm_p_log:.6f}')
print('Conclusion:', 'Homocedasticidad OK' if lm_p_log > 0.05 else 'Persiste heterocedasticidad')
```

```python
# CELDA 15F - Comparacion modelo original vs modelo con log
# 🎯 Objetivo: comparar original vs log(Y) en escala original
comparacion_log = pd.DataFrame({
    'Modelo': ['Original', 'Con log(Y)'],
    'R2 (escala original)': [round(r2, 4), round(r2_log_orig, 4)],
    'RMSE (escala original)': [round(rmse, 6), round(rmse_log_orig, 6)],
    'MAE (escala original)': [round(mae, 6), round(mae_log_orig, 6)],
    'Breusch-Pagan p-value': [round(lm_p, 6), round(lm_p_log, 6)]
})

print(comparacion_log.to_string(index=False))
```

```python
# CELDA 15G - Comparacion homocedasticidad (antes vs log)
# 🎯 Objetivo: comparar visualmente residuos original vs log(Y)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].scatter(modelo_ols_simple.fittedvalues, modelo_ols_simple.resid, alpha=0.3, color='#4ecdc4')
axes[0].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_orig = sm.nonparametric.lowess(modelo_ols_simple.resid, modelo_ols_simple.fittedvalues, frac=0.3)
axes[0].plot(lowess_orig[:, 0], lowess_orig[:, 1], color='#ffa07a', linewidth=2.5)
axes[0].set_title('Modelo original', fontweight='bold')
axes[0].set_xlabel('Valores ajustados')
axes[0].set_ylabel('Residuos')
axes[0].grid(alpha=0.3)

axes[1].scatter(modelo_ols_log.fittedvalues, modelo_ols_log.resid, alpha=0.3, color='#4ecdc4')
axes[1].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_log = sm.nonparametric.lowess(modelo_ols_log.resid, modelo_ols_log.fittedvalues, frac=0.3)
axes[1].plot(lowess_log[:, 0], lowess_log[:, 1], color='#ffa07a', linewidth=2.5)
axes[1].set_title('Modelo con log(Y)', fontweight='bold')
axes[1].set_xlabel('Valores ajustados')
axes[1].set_ylabel('Residuos')
axes[1].grid(alpha=0.3)

plt.suptitle('Comparacion de homocedasticidad: original vs log(Y)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../data/graficas/comparacion_homo_log_dogs.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CELDA 15H - Transformacion raiz cuadrada de la variable objetivo
# 🎯 Objetivo: crear transformacion raiz cuadrada de la variable objetivo
df['y_sqrt'] = np.sqrt(df['vida_promedio'])

print('OK - Se creo y_sqrt = sqrt(vida_promedio)')
print(df[['vida_promedio', 'y_sqrt']].head())
```

```python
# CELDA 15I - Regresion simple con transformacion de raiz
# 🎯 Objetivo: entrenar regresion simple usando sqrt(Y)
X_sqrt = df[['peso_promedio']].values
y_sqrt = df['y_sqrt'].values

X_train_sqrt, X_test_sqrt, y_train_sqrt, y_test_sqrt = train_test_split(
    X_sqrt, y_sqrt, test_size=0.20, random_state=42
)

modelo_sqrt = LinearRegression()
modelo_sqrt.fit(X_train_sqrt, y_train_sqrt)
y_pred_sqrt = modelo_sqrt.predict(X_test_sqrt)

print(f'Intercepto b0 (sqrt): {modelo_sqrt.intercept_:.4f}')
print(f'Coeficiente b1 (sqrt): {modelo_sqrt.coef_[0]:.4f}')
print(f'Ecuacion: sqrt(vida_promedio) = {modelo_sqrt.intercept_:.3f} + ({modelo_sqrt.coef_[0]:.3f}) * peso_promedio')
```

```python
# CELDA 15J - Metricas del modelo con raiz
# 🎯 Objetivo: evaluar metricas sqrt en escala transformada y original
# Metricas en escala transformada (sqrt)
r2_sqrt = r2_score(y_test_sqrt, y_pred_sqrt)
mse_sqrt = mean_squared_error(y_test_sqrt, y_pred_sqrt)
rmse_sqrt = np.sqrt(mse_sqrt)
mae_sqrt = mean_absolute_error(y_test_sqrt, y_pred_sqrt)

# Metricas en escala original de vida_promedio (comparables con modelo original)
y_test_sqrt_orig = np.square(y_test_sqrt)
y_pred_sqrt_orig = np.square(np.maximum(y_pred_sqrt, 0))
r2_sqrt_orig = r2_score(y_test_sqrt_orig, y_pred_sqrt_orig)
rmse_sqrt_orig = np.sqrt(mean_squared_error(y_test_sqrt_orig, y_pred_sqrt_orig))
mae_sqrt_orig = mean_absolute_error(y_test_sqrt_orig, y_pred_sqrt_orig)

print('=' * 58)
print('METRICAS - REGRESION SIMPLE CON RAIZ(Y)')
print('=' * 58)
print('Escala sqrt(Y):')
print(f'R2   : {r2_sqrt:.4f} ({r2_sqrt*100:.2f}%)')
print(f'MSE  : {mse_sqrt:.6f}')
print(f'RMSE : {rmse_sqrt:.6f}')
print(f'MAE  : {mae_sqrt:.6f}')
print()
print('Escala original de vida_promedio (comparable):')
print(f'R2   : {r2_sqrt_orig:.4f} ({r2_sqrt_orig*100:.2f}%)')
print(f'RMSE : {rmse_sqrt_orig:.6f}')
print(f'MAE  : {mae_sqrt_orig:.6f}')
print('=' * 58)
```

```python
# CELDA 15K - OLS del modelo con raiz
# 🎯 Objetivo: obtener resumen OLS del modelo con sqrt(Y)
X_ols_sqrt = sm.add_constant(df['peso_promedio'])
modelo_ols_sqrt = sm.OLS(df['y_sqrt'], X_ols_sqrt).fit()

print(modelo_ols_sqrt.summary())
```

```python
# CELDA 15L - Diagnostico del modelo con raiz
# 🎯 Objetivo: diagnosticar normalidad y homocedasticidad con sqrt(Y)
y_pred_all_sqrt = modelo_sqrt.predict(df[['peso_promedio']].values)
residuos_sqrt = df['y_sqrt'].values - y_pred_all_sqrt

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Diagnosticos - Modelo con raiz(Y)', fontsize=13, fontweight='bold')

axes[0].hist(residuos_sqrt, bins=40, color='#4ecdc4', edgecolor='white', density=True, alpha=0.85)
xr = np.linspace(residuos_sqrt.min(), residuos_sqrt.max(), 200)
axes[0].plot(xr, stats.norm.pdf(xr, residuos_sqrt.mean(), residuos_sqrt.std()), color='#ff6b6b', linewidth=2.5)
axes[0].axvline(0, color='#2a5298', linestyle='--')
axes[0].set_title('Distribucion de residuos', fontweight='bold')
axes[0].set_xlabel('Residuos')
axes[0].set_ylabel('Densidad')
axes[0].grid(axis='y', alpha=0.3)

(osm, osr), (s, i, _) = stats.probplot(residuos_sqrt, dist='norm')
axes[1].scatter(osm, osr, alpha=0.35, s=12, color='#4ecdc4')
axes[1].plot(osm, s * np.array(osm) + i, color='#ff6b6b', linewidth=2.5)
axes[1].set_title('Q-Q plot de residuos', fontweight='bold')
axes[1].set_xlabel('Cuantiles teoricos')
axes[1].set_ylabel('Cuantiles observados')
axes[1].grid(alpha=0.3)

axes[2].scatter(modelo_ols_sqrt.fittedvalues, modelo_ols_sqrt.resid, alpha=0.30, s=15, color='#4ecdc4')
axes[2].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_sqrt = sm.nonparametric.lowess(modelo_ols_sqrt.resid, modelo_ols_sqrt.fittedvalues, frac=0.3)
axes[2].plot(lowess_sqrt[:, 0], lowess_sqrt[:, 1], color='#ffa07a', linewidth=2.5, label='LOWESS')
axes[2].set_title('Residuos vs valores ajustados', fontweight='bold')
axes[2].set_xlabel('Valores ajustados')
axes[2].set_ylabel('Residuos')
axes[2].legend()
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../data/graficas/diagnosticos_sqrt_simple_dogs.png', dpi=150, bbox_inches='tight')
plt.show()

muestra_sw_sqrt = residuos_sqrt if len(residuos_sqrt) <= 5000 else np.random.default_rng(42).choice(residuos_sqrt, 5000, replace=False)
stat_sw_sqrt, p_sw_sqrt = stats.shapiro(muestra_sw_sqrt)
lm_sqrt, lm_p_sqrt, _, _ = het_breuschpagan(modelo_ols_sqrt.resid, modelo_ols_sqrt.model.exog)

print('Shapiro-Wilk (sqrt):')
print(f'p-value: {p_sw_sqrt:.6f}')
print('Conclusion:', 'No se rechaza normalidad' if p_sw_sqrt > 0.05 else 'Se rechaza normalidad')
print()
print('Breusch-Pagan (sqrt):')
print(f'p-value: {lm_p_sqrt:.6f}')
print('Conclusion:', 'Homocedasticidad OK' if lm_p_sqrt > 0.05 else 'Persiste heterocedasticidad')
```

```python
# CELDA 15M - Comparacion modelo original vs modelo con raiz
# 🎯 Objetivo: comparar original vs sqrt(Y) en escala original
comparacion_sqrt = pd.DataFrame({
    'Modelo': ['Original', 'Con sqrt(Y)'],
    'R2 (escala original)': [round(r2, 4), round(r2_sqrt_orig, 4)],
    'RMSE (escala original)': [round(rmse, 6), round(rmse_sqrt_orig, 6)],
    'MAE (escala original)': [round(mae, 6), round(mae_sqrt_orig, 6)],
    'Breusch-Pagan p-value': [round(lm_p, 6), round(lm_p_sqrt, 6)]
})

print(comparacion_sqrt.to_string(index=False))
```

```python
# CELDA 15N - Comparacion homocedasticidad (antes vs raiz)
# 🎯 Objetivo: comparar visualmente residuos original vs sqrt(Y)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].scatter(modelo_ols_simple.fittedvalues, modelo_ols_simple.resid, alpha=0.3, color='#4ecdc4')
axes[0].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_orig = sm.nonparametric.lowess(modelo_ols_simple.resid, modelo_ols_simple.fittedvalues, frac=0.3)
axes[0].plot(lowess_orig[:, 0], lowess_orig[:, 1], color='#ffa07a', linewidth=2.5)
axes[0].set_title('Modelo original', fontweight='bold')
axes[0].set_xlabel('Valores ajustados')
axes[0].set_ylabel('Residuos')
axes[0].grid(alpha=0.3)

axes[1].scatter(modelo_ols_sqrt.fittedvalues, modelo_ols_sqrt.resid, alpha=0.3, color='#4ecdc4')
axes[1].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_sqrt = sm.nonparametric.lowess(modelo_ols_sqrt.resid, modelo_ols_sqrt.fittedvalues, frac=0.3)
axes[1].plot(lowess_sqrt[:, 0], lowess_sqrt[:, 1], color='#ffa07a', linewidth=2.5)
axes[1].set_title('Modelo con sqrt(Y)', fontweight='bold')
axes[1].set_xlabel('Valores ajustados')
axes[1].set_ylabel('Residuos')
axes[1].grid(alpha=0.3)

plt.suptitle('Comparacion de homocedasticidad: original vs sqrt(Y)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../data/graficas/comparacion_homo_sqrt_dogs.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CELDA 15O - Comparacion global: original vs log vs sqrt
# 🎯 Objetivo: comparar globalmente original, log(Y) y sqrt(Y)
comparacion_transformaciones = pd.DataFrame({
    'Modelo': ['Original', 'Con log(Y)', 'Con sqrt(Y)'],
    'R2 (escala original)': [round(r2, 4), round(r2_log_orig, 4), round(r2_sqrt_orig, 4)],
    'RMSE (escala original)': [round(rmse, 6), round(rmse_log_orig, 6), round(rmse_sqrt_orig, 6)],
    'MAE (escala original)': [round(mae, 6), round(mae_log_orig, 6), round(mae_sqrt_orig, 6)],
    'BP p-value': [round(lm_p, 6), round(lm_p_log, 6), round(lm_p_sqrt, 6)]
})

print('🧪 COMPARACION GLOBAL - ORIGINAL vs LOG(Y) vs SQRT(Y)')
print('=' * 94)
print(comparacion_transformaciones.to_string(index=False))
print('=' * 94)
print()
print('✅ Comparacion valida: errores y R2 en escala original de vida_promedio.')
print('📌 Mejor candidato para heterocedasticidad: mayor p-value en Breusch-Pagan.')
```

```python
# CELDA 15P - Comparacion visual triple de homocedasticidad
# 🎯 Objetivo: ver en un bloque los tres graficos de homocedasticidad
fig, axes = plt.subplots(1, 3, figsize=(22, 6))

# Original
axes[0].scatter(modelo_ols_simple.fittedvalues, modelo_ols_simple.resid, alpha=0.30, s=14, color='#4ecdc4')
axes[0].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_orig = sm.nonparametric.lowess(modelo_ols_simple.resid, modelo_ols_simple.fittedvalues, frac=0.3)
axes[0].plot(lowess_orig[:, 0], lowess_orig[:, 1], color='#ffa07a', linewidth=2.5)
axes[0].set_title('Original', fontweight='bold')
axes[0].set_xlabel('Valores ajustados')
axes[0].set_ylabel('Residuos')
axes[0].grid(alpha=0.3)

# Log(Y)
axes[1].scatter(modelo_ols_log.fittedvalues, modelo_ols_log.resid, alpha=0.30, s=14, color='#4ecdc4')
axes[1].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_log = sm.nonparametric.lowess(modelo_ols_log.resid, modelo_ols_log.fittedvalues, frac=0.3)
axes[1].plot(lowess_log[:, 0], lowess_log[:, 1], color='#ffa07a', linewidth=2.5)
axes[1].set_title('Con log(Y)', fontweight='bold')
axes[1].set_xlabel('Valores ajustados')
axes[1].set_ylabel('Residuos')
axes[1].grid(alpha=0.3)

# Sqrt(Y)
axes[2].scatter(modelo_ols_sqrt.fittedvalues, modelo_ols_sqrt.resid, alpha=0.30, s=14, color='#4ecdc4')
axes[2].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_sqrt = sm.nonparametric.lowess(modelo_ols_sqrt.resid, modelo_ols_sqrt.fittedvalues, frac=0.3)
axes[2].plot(lowess_sqrt[:, 0], lowess_sqrt[:, 1], color='#ffa07a', linewidth=2.5)
axes[2].set_title('Con sqrt(Y)', fontweight='bold')
axes[2].set_xlabel('Valores ajustados')
axes[2].set_ylabel('Residuos')
axes[2].grid(alpha=0.3)

plt.suptitle('Comparacion de homocedasticidad: Original vs log(Y) vs sqrt(Y)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../data/graficas/comparacion_homo_triple_dogs.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CELDA 15Q - Mini conclusion automatica (transformaciones)
# 🎯 Objetivo: generar recomendacion automatica de transformacion
ranking_bp = sorted([
    ('Original', lm_p),
    ('log(Y)', lm_p_log),
    ('sqrt(Y)', lm_p_sqrt)
], key=lambda x: x[1], reverse=True)

ranking_rmse = sorted([
    ('Original', rmse),
    ('log(Y)', rmse_log_orig),
    ('sqrt(Y)', rmse_sqrt_orig)
], key=lambda x: x[1])

mejor_bp, p_mejor_bp = ranking_bp[0]
mejor_rmse, val_mejor_rmse = ranking_rmse[0]

print('🧾 MINI CONCLUSION AUTOMATICA')
print('=' * 70)
print(f'🔍 Mejor por heterocedasticidad (Breusch-Pagan): {mejor_bp} (p-value={p_mejor_bp:.6f})')
print(f'🎯 Mejor por error predictivo (RMSE escala original): {mejor_rmse} (RMSE={val_mejor_rmse:.6f})')

if mejor_bp == mejor_rmse:
    print(f'✅ Recomendacion final: usar {mejor_bp}, mejora supuestos y mantiene mejor error.')
else:
    print('⚖️ Recomendacion final: prioriza según objetivo:')
    print(f'   - Inferencia estadistica: {mejor_bp}')
    print(f'   - Minimo error predictivo: {mejor_rmse}')
print('=' * 70)
```

```python
# CELDA 16 - Preparacion y entrenamiento modelo multiple
# 🎯 Objetivo: entrenar modelo de regresion lineal multiple
# Variable booleana como feature numerica (True=1, False=0, None->mediana)
df['hipo_num'] = df['hipoalergenico'].map({True: 1, False: 0})
df['hipo_num'] = df['hipo_num'].fillna(df['hipo_num'].median())

FEATURES = ['peso_promedio', 'diferencia_sexual_peso', 'hipo_num']

X_multi = df[FEATURES].values
y = df['vida_promedio'].values

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X_multi, y, test_size=0.20, random_state=42
)

modelo_multi = LinearRegression()
modelo_multi.fit(X_train_m, y_train_m)

y_pred_multi = modelo_multi.predict(X_test_m)

print(f'Intercepto  b0 : {modelo_multi.intercept_:.4f}')
for feat, coef in zip(FEATURES, modelo_multi.coef_):
    print(f'Coeficiente {feat:>24}: {coef:.4f}')
```

```python
# CELDA 17 - Grafica de coeficientes (multiple)
# 🎯 Objetivo: visualizar e interpretar coeficientes del modelo multiple
colores = ['#ff6b6b' if c > 0 else '#4ecdc4' for c in modelo_multi.coef_]

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(FEATURES, modelo_multi.coef_, color=colores, edgecolor='white', width=0.5)
ax.axhline(0, color='gray', linestyle='--', linewidth=1)
ax.set_title('Coeficientes - Regresion Lineal Multiple', fontsize=12, fontweight='bold')
ax.set_ylabel('Valor del coeficiente')

for bar, coef in zip(bars, modelo_multi.coef_):
    offset = 0.005 if coef >= 0 else -0.01
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + offset, f'{coef:.3f}', ha='center', va='bottom', fontsize=11, fontweight='bold')

ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../data/graficas/coeficientes_multi_dogs.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CELDA 18 - Metricas modelo multiple
# 🎯 Objetivo: calcular metricas de rendimiento del modelo multiple
r2_m = r2_score(y_test_m, y_pred_multi)
mse_m = mean_squared_error(y_test_m, y_pred_multi)
rmse_m = np.sqrt(mse_m)
mae_m = mean_absolute_error(y_test_m, y_pred_multi)

print('=' * 52)
print('📈 METRICAS - REGRESION LINEAL MULTIPLE')
print('=' * 52)
print(f'🎯 R2   : {r2_m:.4f} ({r2_m*100:.2f}%)')
print(f'📉 MSE  : {mse_m:.4f}')
print(f'📏 RMSE : {rmse_m:.4f}')
print(f'🧮 MAE  : {mae_m:.4f}')
print('=' * 52)
```

```python
# CELDA 19 - Resumen OLS multiple
# 🎯 Objetivo: obtener inferencia estadistica OLS del modelo multiple
X_ols_m = sm.add_constant(df[FEATURES])
modelo_ols_multi = sm.OLS(df['vida_promedio'], X_ols_m).fit()
print(modelo_ols_multi.summary())
```

```python
# CELDA 20 - VIF multicolinealidad
# 🎯 Objetivo: medir multicolinealidad mediante VIF
X_vif = df[FEATURES].copy()

vif_data = pd.DataFrame({
    'Variable': FEATURES,
    'VIF': [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
})
vif_data['VIF'] = vif_data['VIF'].round(3)
vif_data['Criterio'] = vif_data['VIF'].apply(
    lambda v: '✅ Sin multicolinealidad (VIF < 5)' if v < 5
    else ('⚠️ Moderada (5 <= VIF < 10)' if v < 10 else '❌ Severa (VIF >= 10)')
)

print('🔎 FACTOR DE INFLACIÓN DE LA VARIANZA (VIF)')
print('=' * 70)
print(vif_data.to_string(index=False))
print('=' * 70)
```

```python
# CELDA 21 - Diagnosticos completos modelo multiple
# 🎯 Objetivo: evaluar diagnosticos del modelo multiple
residuos_m = modelo_ols_multi.resid.values
ajustados_m = modelo_ols_multi.fittedvalues.values

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Diagnosticos - Regresion Lineal Multiple', fontsize=13, fontweight='bold')

axes[0].hist(residuos_m, bins=40, color='#4ecdc4', edgecolor='white', density=True, alpha=0.85)
xr = np.linspace(residuos_m.min(), residuos_m.max(), 200)
axes[0].plot(xr, stats.norm.pdf(xr, residuos_m.mean(), residuos_m.std()), color='#ff6b6b', linewidth=2.5)
axes[0].axvline(0, color='#2a5298', linestyle='--')
axes[0].set_title('Distribucion de residuos', fontweight='bold')
axes[0].set_xlabel('Residuos')
axes[0].set_ylabel('Densidad')
axes[0].grid(axis='y', alpha=0.3)

(osm, osr), (s, i, _) = stats.probplot(residuos_m, dist='norm')
axes[1].scatter(osm, osr, alpha=0.35, s=12, color='#4ecdc4')
axes[1].plot(osm, s * np.array(osm) + i, color='#ff6b6b', linewidth=2.5)
axes[1].set_title('Q-Q plot de residuos', fontweight='bold')
axes[1].set_xlabel('Cuantiles teoricos')
axes[1].set_ylabel('Cuantiles observados')
axes[1].grid(alpha=0.3)

axes[2].scatter(ajustados_m, residuos_m, alpha=0.30, s=15, color='#4ecdc4')
axes[2].axhline(0, color='#ff6b6b', linestyle='--', linewidth=2)
lowess_m = sm.nonparametric.lowess(residuos_m, ajustados_m, frac=0.3)
axes[2].plot(lowess_m[:, 0], lowess_m[:, 1], color='#ffa07a', linewidth=2.5, label='LOWESS')
axes[2].set_title('Residuos vs valores ajustados', fontweight='bold')
axes[2].set_xlabel('Valores ajustados (y-hat)')
axes[2].set_ylabel('Residuos')
axes[2].legend()
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('../data/graficas/diagnosticos_multi_dogs.png', dpi=150, bbox_inches='tight')
plt.show()

muestra_m = residuos_m if len(residuos_m) <= 5000 else np.random.default_rng(42).choice(residuos_m, 5000, replace=False)
stat_sw_m, p_sw_m = stats.shapiro(muestra_m)
lm_m, lm_p_m, _, _ = het_breuschpagan(modelo_ols_multi.resid, modelo_ols_multi.model.exog)

print()
print('Test Shapiro-Wilk (normalidad):')
print(f'p-value: {p_sw_m:.6f}')
print('Conclusion:', 'No se rechaza normalidad' if p_sw_m > 0.05 else 'Se rechaza normalidad')

print()
print('Test Breusch-Pagan (homocedasticidad):')
print(f'p-value: {lm_p_m:.6f}')
print('Conclusion:', 'Homocedasticidad OK' if lm_p_m > 0.05 else 'Heterocedasticidad detectada')
```

```python
# CELDA 22 - Comparacion final de modelos
# 🎯 Objetivo: comparar modelo simple vs multiple con metricas clave
comparacion = pd.DataFrame({
    'Modelo': ['Regresion Simple', 'Regresion Multiple'],
    'Variables': [
        'peso_promedio',
        'peso_promedio + diferencia_sexual_peso + hipo_num'
    ],
    'R2': [round(r2, 4), round(r2_m, 4)],
    'R2 Aj.': [round(modelo_ols_simple.rsquared_adj, 4), round(modelo_ols_multi.rsquared_adj, 4)],
    'RMSE': [round(rmse, 4), round(rmse_m, 4)],
    'MAE': [round(mae, 4), round(mae_m, 4)],
    'AIC': [round(modelo_ols_simple.aic, 2), round(modelo_ols_multi.aic, 2)],
    'BIC': [round(modelo_ols_simple.bic, 2), round(modelo_ols_multi.bic, 2)],
})

print('📊 COMPARACION DE MODELOS - DATA-DOGS')
print('=' * 94)
print(comparacion.to_string(index=False))
print('=' * 94)
print()
print('📌 AIC/BIC más bajo -> mejor equilibrio entre ajuste y parsimonia.')
```
